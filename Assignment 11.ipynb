```python
dataset1=dataset.loc[dataset['Attrition']==1]
dataset1
```

```python
dataset1['TotalWorkingYears'].describe()
```

```python
pop_mean = dataset1['TotalWorkingYears'].mean()
print(pop_mean)
```

```python
#sample_population

sample_data=dataset1.sample(n=20)
sam_mean = sample_data['TotalWorkingYears'].mean()
print("Sample Mean is",sam_mean)
sam_std = sample_data['TotalWorkingYears'].std()
print("Standard Deviation of sample", sam_std)
```

```python
from scipy import stats
## The average total working years of attrition employee is 8.25  years

# To determine this as true a sample of 20 attritioned employee's details are taken. The mean is 12.55 and 
# on standard deviation of  11.20

#Null Hypothesis:  The average total working years of attrition employee is equal to 8.25  years
#Alternative Hypothesis:  The average total working years of attrition employee is not equal to 8.25  years
#print("Enter level of significance")
#alpha = int(input())  #levelOfSignificance

alpha = 0.05
n = 20
dof = n-1
print("Degrees of freedom",dof)

z_test = (sam_mean-pop_mean)/(sam_std/n**0.5)
print (z_test)

z_score = (stats.t.ppf(1-(alpha/2), dof)) #to find table value for two tail test 
print (z_score)                                            # stats.t.ppf(1-0.05, 999) for one tail
if(z_test>-z_score and z_test<z_score):
    print ("Null Hypothesis is accepted")
else:
    print ("Null Hypothesis is regected")
```
