```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
```

```python
dataframe = pd.read_csv("E:\\python\\general_data.csv")
dataframe
```

```python
newdf = dataframe.dropna()
#dataframe
#print(dataframe.isnull().sum())
print(newdf.isnull().sum())
```

```python
newdf.duplicated().sum()
```

```python
newdf.drop_duplicates()
```

```python
newdf["Attrition"].replace({"Yes":1,"No":0},inplace=True)
```

```python
newdf["Attrition"].value_counts()
```

```python
#print(newdf["StandardHours"].isnull().sum())

newdf.corr()
```

### Correlation between Attrition and Age

```python
stats,p= pearsonr(newdf.Attrition,newdf.Age)

print("Correlation value is",stats)
print("p-value is",p)

if p<0.05:
    print("Null Hypothesis is regected")
else:
    print("Null Hypothesis is accepted")

plt.scatter(newdf.Attrition,newdf.Age)
```

### Correlation between YearsAtCompany and Attrition
